## Importando bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Carregando os dados

```python
# Carregando dados
df = pd.read_csv('../data/raw/train.csv', parse_dates=['Date'])
```

```python
# Vis√£o geral
df.head()
```

```python
# Verificando valores nulos
df.isnull().sum()
```

## Processamento

```python
# Selecionando uma loja para an√°lise inicial
store_id = 1
store_df = df[df['Store'] == store_id]

# Gr√°fico de vendas ao longo do tempo
plt.figure(figsize=(12, 4))
plt.plot(store_df['Date'], store_df['Sales'])
plt.title(f'Vendas da Loja {store_id} ao longo do tempo')
plt.xlabel('Data')
plt.ylabel('Vendas')
plt.grid()
plt.show()
```
